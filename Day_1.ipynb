```python
# Graph Drawing
N = 100
dx = 2*np.pi/N
x_arr = []
y_arr = []
for i in range(N) :
  x = i*dx
  y = np.sin(x)
  x_arr.append(x)
  y_arr.append(y)
plt.plot(x_arr,y_arr, 'o')
plt.plot(x_arr,y_arr)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
# Colorbar

x = np.random.normal(0, 1, 10)
y = np.random.normal(0, 1, 10)
sizes = np.random.uniform(10, 100, 10)
colors = np.random.uniform(0, 1, 10)
plt.scatter(x, y, s=sizes, c=colors)
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.show()
```

```python
#bar graph

x = np.arange(7) + 0.5
y = np.random.uniform(1, 9, len(x))
plt.bar(x, y, width=1, edgecolor='white')
plt.xlabel('days')
plt.ylabel('hours')
plt.show()
```

```python
# Histogram

x = np.random.normal(0,1,10000)
plt.hist(x, bins=15, edgecolor='white')
plt.ylabel('frequency')
plt.show()
```
